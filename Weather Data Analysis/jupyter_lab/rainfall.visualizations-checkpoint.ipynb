```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load Data
df = pd.read_csv("weather_data.csv")

# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Identify Dangerous Rainfall Days (Rainfall > 30mm)
dangerous_rainfall_days = df["Rainfall (mm)"] > 30

# Create Plot
plt.figure(figsize=(12, 5))

# Plot Rainfall
plt.plot(df["Date"], df["Rainfall (mm)"], marker="o", linestyle="-", 
         color="blue", alpha=0.6, label="Rainfall (mm)")

# Highlight Dangerous Rainfall Days in Red
plt.scatter(df["Date"][dangerous_rainfall_days], df["Rainfall (mm)"][dangerous_rainfall_days], 
            color="red", label="Dangerous Rainfall (>30mm)", zorder=3)

# Labels & Titles
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.title("🌧️ Rainfall Trends Over 6 Months")

# Improve X-axis Date Formatting
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)

# Add Legend Outside to Prevent Overlap
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

# Save the Plot (Optional)
plt.savefig("rainfall_plot.png", dpi=300, bbox_inches='tight')

# Show Plot
plt.show()
```
